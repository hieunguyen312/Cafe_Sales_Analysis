```python
df = df.drop_duplicates()
len(df)
# No duplicates
```

```python
# Create an unique price dict for all items on the menu
unique_price = {'Coffee' : 2.0, 'Tea' : 1.5, 'Sandwich' : 4.0, 'Salad' :5.0, 'Cake' : 3.0,\
'Cookie' : 1.0, 'Smoothie' : 4.0, 'Juice' : 3.0}
print(unique_price)
```

```python
# Format Item, Quantity and Price Per Unit cols to numeric
df[["Total Spent", "Quantity", "Price Per Unit"]] = df[["Total Spent", "Quantity", "Price Per Unit"]].apply(pd.to_numeric, errors = "coerce", axis = 1)
df
```

```python
# Populate Price Per Unit col using unique price, matching the value with Item col
df["Price Per Unit"] = df.apply(lambda row: unique_price.get(row["Item"]) if pd.isna(row["Price Per Unit"]) else row["Price Per Unit"], axis = 1)
df
```

```python
# Populating Price Per Unit col using Total Spent / Quantity
df["Price Per Unit"] = df.apply(lambda row: row["Total Spent"] / row["Quantity"] if pd.isna(row["Price Per Unit"]) else row["Price Per Unit"], axis = 1)
df
```

```python
df.isna().sum()
```

```python
# Populate Total Spent and Quantity col
df["Total Spent"] = df.apply(lambda row: row["Price Per Unit"] * row["Quantity"] if pd.isna(row["Total Spent"]) else row["Total Spent"], axis = 1)
df.isna().sum()
df
```

```python
# Populate Quantity col
df["Quantity"] = df.apply(lambda row: row["Total Spent"] / row["Price Per Unit"] if pd.isna(row["Quantity"]) else row["Quantity"], axis = 1)
df.isna().sum()
df
```

```python
# drop null columns
df.dropna(subset =["Price Per Unit"], inplace = True)
df.isna().sum()
df
```

```python
# Populate Quantity and Total Spent with mean value
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df["Total Spent"] = df["Total Spent"].fillna(df["Total Spent"].mean())
df.isna().sum()
df
```

```python
# Check for possible null indicators
selected_cols = ["Transaction ID", "Item", "Payment Method","Location","Transaction Date"]

value_columns = {}

for column in selected_cols:
    for value in df[column].unique():
        if value not in value_columns:
            value_columns[value] = []
        value_columns[value].append(column)

# Convert to df
value_columns_df = pd.DataFrame([(value, cols, len(cols)) for value, cols in value_columns.items()], 
                                columns=["Value", "Columns", "Column Count"])

# Filter values appearing in multiple columns
suspected_nulls = value_columns_df[value_columns_df["Column Count"] > 1]

print(suspected_nulls)
# UNKOWN and ERRORS are used along side with NaN as null values
df
```

```python
# Replace null in Item to np.nan
df["Item"] = df["Item"].replace({"UNKNOWN" : np.nan, "ERROR" : np.nan})
df.isnull().sum()

# Repopulate Item based on unduplicated price based on value in Price Per Unit
non_dup_prices = {"Cookie": 1.0, "Tea": 1.5, "Coffee": 2.0, "Salad": 5.0}
df["Item"] = df.apply(lambda row: next((k for k, v in non_dup_prices.items() if v == row["Price Per Unit"]), row["Item"]), axis = 1)
df.isnull().sum()
df
```

```python
# Replace null values with np.nan
df["Payment Method"] = df["Payment Method"].replace({"UNKNOWN" : np.nan, "ERROR" : np.nan})
df["Location"] = df["Location"].replace({"UNKNOWN" : np.nan, "ERROR" : np.nan})
df.isnull().sum()
df
```

```python
# Format Transaction Date col
# Make Month Name, Month No and Year col for analysis uses
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors = "coerce")
df["Year"] = pd.to_datetime(df["Transaction Date"]).dt.strftime("%Y")
df["Month Name"] = pd.to_datetime(df["Transaction Date"]).dt.strftime("%b")
df["Month Name"] = pd.to_datetime(df["Transaction Date"]).dt.strftime("%m")
df.head(21)
```
